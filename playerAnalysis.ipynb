```python
# return top 10 players with most fantasy points
top_10_most_fantasy_points = df.sort_values(by=["FantasyPoints"], ascending=False).head(10)
top_10_most_fantasy_points
```

```python
# return top 10 players with the most assists
top_10_most_assists = df.sort_values(by=["AST"], ascending=False).head(10)
top_10_most_assists
```
